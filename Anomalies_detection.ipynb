```python
# Criptos recomendadas 

# BTCUSDT → Bitcoin
# ETHUSDT → Ethereum
# BNBUSDT → Binance Coin
# ADAUSDT → Cardano
# SOLUSDT → Solana
# XRPUSDT → Ripple
```

```python
# 1. Variables temporales

# open_time
# Fecha y hora (UTC) en la que inicia el intervalo de 15 minutos.
# Cada fila representa lo ocurrido a partir de este instante.

# close_time
# Fecha y hora (UTC) en la que finaliza el intervalo de 15 minutos.
# Marca el cierre exacto de la vela.

# 2. Variables de precio (OHLC)

# open
# Precio de apertura del activo al inicio del intervalo.

# high
# Precio máximo alcanzado durante el intervalo.

# low
# Precio mínimo alcanzado durante el intervalo.

# close
# Precio de cierre del activo al final del intervalo.

# Estas cuatro variables resumen el comportamiento del precio dentro del periodo.

# 3. Variables de volumen y actividad

# volume
# Cantidad total del activo base negociado durante el intervalo
# (en este caso, Bitcoin).

# quote_asset_volume
# Volumen total negociado expresado en la moneda cotizada
# (en este caso, USDT).

# n_trades
# Número total de operaciones (trades) ejecutadas durante el intervalo.
# Refleja el nivel de actividad del mercado.


# 4. Variables de agresividad de mercado

# taker_buy_base
# Cantidad del activo base (BTC) comprada mediante órdenes agresivas
# (órdenes de mercado).

# taker_buy_quote
# Valor en la moneda cotizada (USDT) de las compras agresivas realizadas.

# Estas variables permiten medir la presión compradora del mercado.


# 5. Variables técnicas / administrativas

# ignore
# Campo técnico sin uso analítico.
# Su valor es siempre cero y puede eliminarse sin impacto.

# symbol
# Identificador del activo financiero analizado
# (por ejemplo, BTCUSDT).

# interval
# Intervalo temporal de agregación de los datos.
# En este caso, 15m indica que cada fila resume 15 minutos de mercado.
```

```python
# Los mercados financieros digitales, como el de las criptomonedas, se caracterizan por alta volatilidad, 
# operación continua y rápidos cambios en el comportamiento del precio y el volumen. Estas condiciones favorecen 
# la aparición de eventos atípicos tales como picos abruptos de precio, caídas repentinas, incrementos inusuales de volumen o 
# patrones asociados a liquidaciones masivas.

# La detección temprana de estos eventos es de gran interés para:

# gestión de riesgos,
# monitoreo de estabilidad del mercado,
# análisis de comportamiento financiero.

# Sin embargo, los métodos estadísticos tradicionales suelen ser limitados para capturar patrones no lineales y dinámicos,
# especialmente en datos de alta frecuencia.
```
